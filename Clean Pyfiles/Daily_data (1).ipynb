# Heart Rate Data

Loading data

```python
hr = pd.read_csv('/content/drive/MyDrive/Week 8 (Capstone Project) - Fitbit Dataset/heartrate_seconds_merged.csv')
```

Basic Analysis

```python
hr.head(3)
```

```python
hr.dtypes
```

```python
hr.shape
```

```python
hr.isna().sum()
```

Transforming data

```python
hr['Time'] = pd.to_datetime(hr['Time'], format='%m/%d/%Y %I:%M:%S %p')
hr['Day'] = hr['Time'].dt.date
hr['Time'] = hr['Time'].dt.strftime('%H:%M')
date_column = hr.pop('Day')
hr.insert(1, 'Day', date_column)
```

```python
hr .head(3)
```

```python
hr["Day"] = pd.to_datetime(hr["Day"]).dt.strftime("%d-%m-%Y")
```

```python
hr.head(3)
```

```python
hr['Value'] = hr.groupby(['Id', 'Day', 'Time'])['Value'].transform('median')
hr = hr.drop_duplicates(subset=['Id', 'Day', 'Time'], keep='first')
hr.rename(columns={'Value': 'Heartrate'}, inplace=True)
hr['Heartrate'] = hr['Heartrate'].astype(int)
hr['DailyMeanHeartrate'] = hr.groupby(['Id', 'Day'])['Heartrate'].transform('mean')
```

```python
hr.head(3)
```

```python
hr.duplicated().sum()
```

```python
hr= hr.sort_values(by=['Id', 'Day', 'Time'])
```

```python
hr.to_csv('heart_rate_cleaned.csv', index=False)
```

```python
hr.shape
```

```python
wi.drop(columns=['Fat', 'IsManualReport', 'LogId'], inplace=True)
```

Transforming Data

```python
wi['Date'] = pd.to_datetime(wi['Date'], format='%m/%d/%Y %I:%M:%S %p')
wi['Date Weight'] = wi['Date'].dt.strftime('%Y-%m-%d')
wi.drop(columns=['Date'], inplace=True)
date_weight_column = wi.pop('Date Weight')
wi.insert(1, 'Date Weight', date_weight_column)
```

```python
wi.head(3)
```

```python
wi['WeightKg'] = wi['WeightKg'].round(2)
wi['WeightPounds'] = wi['WeightPounds'].round(2)
wi['BMI'] = wi['BMI'].round(2)
```

```python
melted_df = wi.melt(id_vars=['Id', 'Date Weight'], value_vars=['WeightKg', 'WeightPounds', 'BMI'], var_name='Variable', value_name='Value')
sorted_df1 = melted_df.sort_values(by=['Id', 'Date Weight'])
pivoted_df1 = sorted_df1.pivot_table(index=['Id', 'Date Weight'], columns='Variable', values='Value', aggfunc='first').reset_index()
```

```python
pivoted_df1.head(3)
```

```python
pivoted_df1["Date Weight"] = pd.to_datetime(pivoted_df1["Date Weight"]).dt.strftime("%d-%m-%Y")
```

# Merging Datasets

```python
df1 = pd.read_csv('/content/daily_activity_cleaned.csv')
df2 = pd.read_csv('/content/daily_calories_cleaned.csv')
df3 = pd.read_csv('/content/daily_intensities_cleaned.csv')
df4 = pd.read_csv('/content/daily_steps_cleaned.csv')
df5 = pd.read_csv('/content/heart_rate_cleaned.csv')
df6 = pd.read_csv('/content/sleep_day_data_cleaned.csv')
df7 = pd.read_csv('/content/weight_data_cleaned.csv')
```

```python
df5 = df5.drop(columns=['Heartrate', 'Time'])
```

```python
df5.shape
```

```python
df5.head(3)
```

```python
df5.shape
```

```python
df5 = df5.drop_duplicates()
```

```python
# Merge DataFrames on specified columns
activity_heart = pd.merge(df1, df5, left_on=['Id', 'ActivityDate'], right_on=['Id', 'Day'], how='left')

# Move the 'Date Weight' column
moving_column = activity_heart.pop('DailyMeanHeartrate')
activity_heart.insert(2, 'DailyMeanHeartrate', moving_column)

# Rename the 'Value' column to 'Heartrate'
activity_heart.rename(columns={'DailyMeanHeartrate': 'DailyAverageHeartrate'}, inplace=True)

activity_heart.head(5)
```

```python
activity_heart.shape
```

```python
activity_sleep = pd.merge(activity_heart, df6, left_on=['Id', 'ActivityDate'], right_on=['Id', 'SleepDay'], how='left')
activity_sleep.head(5)
```
