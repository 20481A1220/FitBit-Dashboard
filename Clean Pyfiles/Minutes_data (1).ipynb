 # Merging Datasets

```python
m1 = pd.read_csv('/content/mMET_cleaned.csv')
m2 = pd.read_csv('/content/mcalories_cleaned.csv')
m3 = pd.read_csv('/content/mintensitie_cleaned.csv')
m4 = pd.read_csv('/content/msleep_cleaned.csv')
m5 = pd.read_csv('/content/msteps_cleaned.csv')
```

```python
m = pd.merge(m1, m2, on=['Id', 'ActivityDay', 'ActivityMinute'], how='left')
```

```python
m.head(3)
```

```python
mi = pd.merge(m, m3, on=['Id', 'ActivityDay', 'ActivityMinute'], how='left')
```

```python
mi.head(3)
```

```python
ms = pd.merge(mi, m4, on=['Id', 'ActivityDay', 'ActivityMinute'], how='left')
```

```python
ms.head(3)
```

```python
f3 = pd.merge(ms, m5, on=['Id', 'ActivityDay', 'ActivityMinute'], how='left')
```

```python
f3.head(3)
```

```python
f3.isna().sum()
```

```python
f3['Sleep'] = f3['Sleep'].fillna(0)
f3['Sleep'] = f3['Sleep'].astype(int)
```

```python
f3.to_csv('minute_stats.csv', index=False)
```
