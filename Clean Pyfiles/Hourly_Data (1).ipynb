# Hourly Calories Data

Loading the data

```python
df1 = pd.read_csv('/content/drive/MyDrive/Week 8 (Capstone Project) - Fitbit Dataset/hourlyCalories_merged.csv')
```

Basic Analysis

```python
df1.head(3)
```

```python
df1.shape
```

```python
df1.dtypes
```

```python
df1.describe()
```

```python
df1.isna().sum()
```

Transforming Data

```python
df1['ActivityHour'] = pd.to_datetime(df1['ActivityHour'], format='%m/%d/%Y %I:%M:%S %p')
df1['ActivityDay'] = df1['ActivityHour'].dt.date
df1['ActivityHour'] = df1['ActivityHour'].dt.strftime('%H:%M')
date_column = df1.pop('ActivityDay')
df1.insert(1, 'ActivityDay', date_column)
df1.head(5)
```

```python
df1["ActivityDay"] = pd.to_datetime(df1["ActivityDay"]).dt.strftime("%d-%m-%Y")
```

```python
df1.head(3)
```

```python
df1.duplicated().sum()
```

```python
df1 = df1.sort_values(by=['Id', 'ActivityDay', 'ActivityHour'])
```

```python
df1.to_csv('hcalories_cleaned.csv', index=False)
```

# Merging the Datasets

```python
h1 = pd.read_csv('/content/hcalories_cleaned.csv')
h2 = pd.read_csv('/content/hintensitie_cleaned.csv')
h3 = pd.read_csv('/content/hourly_steps_cleaned.csv')
```

```python
h = pd.merge(h1, h2, on=['Id', 'ActivityDay', 'ActivityHour'], how='left')
```

```python
h.head(3)
```

```python
f2 = pd.merge(h, h3, on=['Id', 'ActivityDay', 'ActivityHour'], how='left')
```

```python
f2.head(3)
```

```python
f2.shape
```

```python
f2.isna().sum()
```

```python
f2.to_csv('hourly_stats.csv', index=False)
```
